```python
import  pandas  as pd
import  numpy  as np
from prettytable import PrettyTable
```

```python
df = pd.read_csv("iris_data.csv",names=['sepal length','sepal width','petal length','petal width','class'], )
df.drop(['petal length','petal width'],axis='columns', inplace=True)

print(df.head(n=6))

print(df['class'].unique())
```

```python
setosa=df[df['class']=='Iris-setosa']
versicolor=df[df['class']=='Iris-versicolor']
virginica=df[df['class']=='Iris-virginica']

N=df['class'].count()
N_setosa=setosa['class'].count()
N_versicolor=versicolor['class'].count()
N_virginica=virginica['class'].count()
```

```python
#used in len_con2 because of fitting the result (100% instead of 100,1% in summary)
def truncate(n, decimals=0):
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier

N1=[setosa['sepal length'].count(), versicolor['sepal length'].count(), virginica['sepal length'].count(), df['sepal length'].count() ]
Mean1=[round(setosa['sepal length'].mean(),1), round(versicolor['sepal length'].mean(),1), round(virginica['sepal length'].mean(),1), round(df['sepal length'].mean(),1) ]
Min1=[setosa['sepal length'].min(), versicolor['sepal length'].min(), virginica['sepal length'].min(), df['sepal length'].min() ]
Median1=[round(setosa['sepal length'].median(),1), round(versicolor['sepal length'].median(),1), round(virginica['sepal length'].median(),1), round(df['sepal length'].median(),1) ]
Max1=[setosa['sepal length'].max(), versicolor['sepal length'].max(), virginica['sepal length'].max(), df['sepal length'].max() ]
Std1=[round(setosa['sepal length'].std(),1), round(versicolor['sepal length'].std(),1), round(virginica['sepal length'].std(),1), round(df['sepal length'].std(),1) ]

N2=[setosa['sepal width'].count(), versicolor['sepal width'].count(), virginica['sepal width'].count(), df['sepal width'].count() ]
Mean2=[round(setosa['sepal width'].mean(),1), round(versicolor['sepal width'].mean(),1), round(virginica['sepal width'].mean(),1), round(df['sepal width'].mean(),1) ]
Min2=[setosa['sepal width'].min(), versicolor['sepal width'].min(), virginica['sepal width'].min(), df['sepal width'].min() ]
Median2=[setosa['sepal width'].median(), versicolor['sepal width'].median(), virginica['sepal width'].median(), df['sepal width'].median() ]
Max2=[setosa['sepal width'].max(), versicolor['sepal width'].max(), virginica['sepal width'].max(), df['sepal width'].max() ]
Std2=[round(setosa['sepal width'].std(),1), round(versicolor['sepal width'].std(),1), round(virginica['sepal width'].std(),1), round(df['sepal width'].std(),1) ]

len_con1=[(setosa['sepal length']<5).sum(), round((setosa['sepal length']<5).sum()/N_setosa*100 ,1),
          (versicolor['sepal length']<5).sum(), round((versicolor['sepal length']<5).sum()/N_versicolor*100,1),
          (virginica['sepal length']<5).sum(), round((virginica['sepal length']<5).sum()/N_virginica*100,1),  
          (df['sepal length']<5).sum(), round((df['sepal length']<5).sum()/N*100,1)
         ]

len_con2 =[((setosa['sepal length']>=5)&(setosa['sepal length']<6)).sum() , round(((setosa['sepal length']>=5)&(setosa['sepal length']<6)).sum()/N_setosa*100,1),
           ((versicolor['sepal length']>=5)&(versicolor['sepal length']<6)).sum() , round(((versicolor['sepal length']>=5)&(versicolor['sepal length']<6)).sum()/N_versicolor*100,1),
           ((virginica['sepal length']>=5)&(virginica['sepal length']<6)).sum() , round(((virginica['sepal length']>=5)&(virginica['sepal length']<6)).sum()/N_virginica*100,1),
           ((df['sepal length']>=5)&(df['sepal length']<6)).sum() , truncate(((df['sepal length']>=5)&(df['sepal length']<6)).sum()/N*100,1)
          ]

len_con3 =[((setosa['sepal length']>=6)&(setosa['sepal length']<7)).sum() , round(((setosa['sepal length']>=6)&(setosa['sepal length']<7)).sum()/N_setosa*100,1),
           ((versicolor['sepal length']>=6)&(versicolor['sepal length']<7)).sum() , round(((versicolor['sepal length']>=6)&(versicolor['sepal length']<7)).sum()/N_versicolor*100,1),
           ((virginica['sepal length']>=6)&(virginica['sepal length']<7)).sum() , round(((virginica['sepal length']>=6)&(virginica['sepal length']<7)).sum()/N_virginica*100,1),
           ((df['sepal length']>=6)&(df['sepal length']<7)).sum() , round(((df['sepal length']>=6)&(df['sepal length']<7)).sum()/N*100,1)
          ]

len_con4=[(setosa['sepal length']>=7).sum(), round((setosa['sepal length']>=7).sum()/N_setosa*100 ,1),
          (versicolor['sepal length']>=7).sum(), round((versicolor['sepal length']>=7).sum()/N_versicolor*100,1),
          (virginica['sepal length']>=7).sum(), round((virginica['sepal length']>=7).sum()/N_virginica*100,1),  
          (df['sepal length']>=7).sum(), (df['sepal length']>=7).sum()/N*100
         ]

wg_con1=[(setosa['sepal width']<3).sum(), round((setosa['sepal width']<3).sum()/N_setosa*100 ,1),
          (versicolor['sepal width']<3).sum(), round((versicolor['sepal width']<3).sum()/N_versicolor*100,1),
          (virginica['sepal width']<3).sum(), round((virginica['sepal width']<3).sum()/N_virginica*100,1),  
          (df['sepal width']<3).sum(), round((df['sepal width']<3).sum()/N*100,1)
         ]

wg_con2 =[((setosa['sepal width']>=3)&(setosa['sepal width']<3.5)).sum() , round(((setosa['sepal width']>=3)&(setosa['sepal width']<3.5)).sum()/N_setosa*100,1),
           ((versicolor['sepal width']>=3)&(versicolor['sepal width']<3.5)).sum() , round(((versicolor['sepal width']>=3)&(versicolor['sepal width']<3.5)).sum()/N_versicolor*100,1),
           ((virginica['sepal width']>=3)&(virginica['sepal width']<3.5)).sum() , round(((virginica['sepal width']>=3)&(virginica['sepal width']<3.5)).sum()/N_virginica*100,1),
           ((df['sepal width']>=3)&(df['sepal width']<3.5)).sum() , round(((df['sepal width']>=3)&(df['sepal width']<3.5)).sum()/N*100,1)
          ]

wg_con3 =[((setosa['sepal width']>=3.5)&(setosa['sepal width']<4)).sum() , round(((setosa['sepal width']>=3.5)&(setosa['sepal width']<4)).sum()/N_setosa*100,1),
           ((versicolor['sepal width']>=3.5)&(versicolor['sepal width']<4)).sum() , round(((versicolor['sepal width']>=3.5)&(versicolor['sepal width']<4)).sum()/N_versicolor*100,1),
           ((virginica['sepal width']>=3.5)&(virginica['sepal width']<4)).sum() , round(((virginica['sepal width']>=3.5)&(virginica['sepal width']<4)).sum()/N_virginica*100,1),
           ((df['sepal width']>=3.5)&(df['sepal width']<4)).sum() , round(((df['sepal width']>=3.5)&(df['sepal width']<4)).sum()/N*100,1)
          ]

```

```python
y = PrettyTable()

y = PrettyTable(["All Flowers", "Iris Setosa", "Iris Versicolor", "Iris Virginica", "Total"] )

y.title = 'Iris Flower Summary'

y.add_row(["\n","N=%d"%(N1[0]), "N=%d"%(N1[1]), "N=%d"%(N1[2]),"N=%d"%(N1[3])])
y.add_row(["SEPAL LENGTH [CM]","", "","",""])
y.add_row(["  N",N1[0], N1[1],N1[2],N1[3]])
y.add_row(["  MEAN",Mean1[0],Mean1[1],Mean1[2],Mean1[3]])
y.add_row(["  MIN",Min1[0],Min1[1],Min1[2],Min1[3]])
y.add_row(["  MEDIAN",Median1[0],Median1[1],Median1[2],Median1[3]])
y.add_row(["  MAX",Max1[0],Max1[1],Max1[2],Max1[3]])
y.add_row(["  STANDARD DEVIATION\n",Std1[0],Std1[1],Std1[2],Std1[3]])

y.add_row(["SEPAL LENGTH (%)","", "","",""])
y.add_row(["  < 5","%d(%.1f)"%(len_con1[0],len_con1[1]), "%d(%.1f)"%(len_con1[2],len_con1[3]),"%d(%.1f)"%(len_con1[4],len_con1[5]),"%d(%.1f)"%(len_con1[6],len_con1[7])])
y.add_row(["  >=5 AND <6","%d(%.1f)"%(len_con2[0],len_con2[1]), "%d(%.1f)"%(len_con2[2],len_con2[3]),"%d(%.1f)"%(len_con2[4],len_con2[5]),"%d(%.1f)"%(len_con2[6],len_con2[7])])
y.add_row(["  >=6 AND <7","%d(%.1f)"%(len_con3[0],len_con3[1]), "%d(%.1f)"%(len_con3[2],len_con3[3]),"%d(%.1f)"%(len_con3[4],len_con3[5]),"%d(%.1f)"%(len_con3[6],len_con3[7])])
y.add_row(["  >= 7\n","%d(%.1f)"%(len_con4[0],len_con4[1]), "%d(%.1f)"%(len_con4[2],len_con4[3]),"%d(%.1f)"%(len_con4[4],len_con4[5]),"%d(%.1f)"%(len_con4[6],len_con4[7])])

y.add_row(["SEPAL WIDTH [CM]","", "","",""])
y.add_row(["  N",N2[0], N2[1],N2[2],N2[3]])
y.add_row(["  MEAN",Mean2[0],Mean2[1],Mean2[2],Mean2[3]])
y.add_row(["  MIN",Min2[0],Min2[1],Min2[2],Min2[3]])
y.add_row(["  MEDIAN",Median2[0],Median2[1],Median2[2],Median2[3]])
y.add_row(["  MAX",Max2[0],Max2[1],Max2[2],Max2[3]])
y.add_row(["  STANDARD DEVIATION\n",Std2[0],Std2[1],Std2[2],Std2[3]])

y.add_row(["SEPAL WIDTH (%)","", "","",""])
y.add_row(["  < 3","%d(%.1f)"%(wg_con1[0],wg_con1[1]), "%d(%.1f)"%(wg_con1[2],wg_con1[3]),"%d(%.1f)"%(wg_con1[4],wg_con1[5]),"%d(%.1f)"%(wg_con1[6],wg_con1[7])])
y.add_row(["  >=3 AND <3.5","%d(%.1f)"%(wg_con2[0],wg_con2[1]), "%d(%.1f)"%(wg_con2[2],wg_con2[3]),"%d(%.1f)"%(wg_con2[4],wg_con2[5]),"%d(%.1f)"%(wg_con2[6],wg_con2[7])])
y.add_row(["  >=3.5 AND <4","%d(%.1f)"%(wg_con3[0],wg_con3[1]), "%d(%.1f)"%(wg_con3[2],wg_con3[3]),"%d(%.1f)"%(wg_con3[4],wg_con3[5]),"%d(%.1f)"%(wg_con3[6],wg_con3[7])])

y.align["All Flowers"] = "l"

print(y.get_string(fields=["All Flowers", "Iris Setosa", "Iris Versicolor", "Iris Virginica", "Total"]))
```

```python
file = open('iris-flowers.txt', 'w')
file.write(y.get_string(fields=["All Flowers", "Iris Setosa", "Iris Versicolor", "Iris Virginica", "Total"]))
file.close()
```
